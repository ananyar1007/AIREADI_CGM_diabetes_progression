```python
import pandas as pd
import numpy as np
import os
base_dir = r'C:\\Users\preet\\Documents\\ai-readi-notebooks\\main\\CGM_diabetes_progression\\results2\\'
```

```python
result_list = os.listdir(base_dir)
```

```python
sub_str_list = [ 
                'Dexcom_CGM_shifted2_Txfm_p3_Label_study_dataset_1_2',
                'Dexcom_CGM_Txfm_p3_Label_study_dataset_1_2',
                'Dexcom_CGM_Txfm_None_Label_study_dataset_1_2',
                'Libre_CGM_shifted2_Txfm_p3_Label_study_dataset_1_2',
                'Libre_CGM_Txfm_p3_Label_study_dataset_1_2',
                'Libre_CGM_Txfm_None_Label_study_dataset_1_2',
                 ]
pd.set_option('display.float_format', '{:.2g}'.format)

for sub_str in sub_str_list:
    matching_file_names = [fname for fname in result_list if sub_str in fname]
    print(sub_str, matching_file_names)
    dfs = []
    for file_name in matching_file_names:
        file_path = os.path.join(base_dir, file_name)
        #print(f"  Loading {file_path}")
        df = pd.read_csv(file_path)
        df.drop(columns=['Unnamed: 0'],inplace=True)
        dfs.append(df)
    

    # Extract numeric part: columns 1 onward
    numeric_arrays = [df.iloc[:, 1:].values for df in dfs]   # each (7,5)
    # Stack -> shape (N,7,5)
    stacked = np.stack(numeric_arrays, axis=0)
    # Element-wise mean -> (7,5)
    avg_numeric = stacked.mean(axis=0)
 # Std dev across dfs: shape (7,5)
    err_numeric = stacked.std(axis=0, ddof=0)   # population std

    # Restore first column
    first_col = dfs[0].iloc[:, 0]

    # ---------- Build Average DF ----------
    avg_df = pd.concat(
        [first_col, pd.DataFrame(avg_numeric, columns=dfs[0].columns[1:])],
        axis=1
    )

    # ---------- Build Error DF ----------
    err_df = pd.concat(
        [first_col, pd.DataFrame(err_numeric, columns=dfs[0].columns[1:])],
        axis=1
    )

    print(f"\nFinal averaged dataframe (shape {avg_df.shape}):")
    print(avg_df)

    print(f"\nError dataframe (std dev, shape {err_df.shape}):")
    print(err_df)
```
